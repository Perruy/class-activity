```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#Here I load and display the image
image_path = "/content/pcb.tif"
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
plt.figure(figsize=(6,6))
plt.imshow(original_image, cmap="gray")
plt.title("Original Grayscale Image")
plt.axis("off")
plt.show()

# secondly:I then identify noise type using histogram analysis
plt.figure(figsize=(6,4))
plt.hist(original_image.ravel(), bins=256, range=(0,256), color="blue", alpha=0.7)
plt.title("Histogram of Image")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()

# Thirdly: I apply appropriate filtering based on noise type

# 1. Median Filter
median_filtered = cv2.medianBlur(original_image, 5)

# 2. Gaussian Filter
gaussian_filtered = cv2.GaussianBlur(original_image, (5,5), 0)

# 3 Bilateral Filter
bilateral_filtered = cv2.bilateralFilter(original_image, 9, 75, 75)

fig, axes = plt.subplots(1, 3, figsize=(12,4))
axes[0].imshow(median_filtered, cmap="gray")
axes[0].set_title("Median Filter (Salt & Pepper)")
axes[0].axis("off")
axes[1].imshow(gaussian_filtered, cmap="gray")
axes[1].set_title("Gaussian Filter (Gaussian Noise)")
axes[1].axis("off")
axes[2].imshow(bilateral_filtered, cmap="gray")
axes[2].set_title("Bilateral Filter (Edge-Preserving)")
axes[2].axis("off")

plt.show()
```
